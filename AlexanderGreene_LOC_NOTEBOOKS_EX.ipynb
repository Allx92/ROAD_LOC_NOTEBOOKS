## Importing a digital collection

The `requests` module allows you to send HTTP requests using Python. Get request to 'Coptic Orthodox Liturgical Chant' collection URL is sent using requests.get() function. The response from the URL is then converted from JSON format to a python dictionary using .json() function.

```python
import requests # Importing 'requests' library
r = requests.get
r = requests.get("https://www.loc.gov/collections/coptic-orthodox-liturgical-chant/?fo=json") # Sending a GET HTTP request to the specified URL
r.json() # Parsing the HTTP response from JSON format to a python dictionary
```

 ## Fetching Data and Iterating through a Digital Collection

  The code makes a GET HTTP request to a specific URL and converts the response from JSON format to a Python dictionary. Then it iterates over each collection in the 'results' list within the dictionary and prints the 'title' attribute of each collection.

```python
collections_json = requests.get("https://www.loc.gov/collections/coptic-orthodox-liturgical-chant/?fo=json").json() #This line uses the requests library to fetch the content of the specified URL

for collection in collections_json["results"]:#This line iterates through each "collection" element within the "results" key of the collections_json dictionary
    print(collection["title"]) #print(collection["title"]): Inside the loop, this line retrieves the "title" value from the current collection dictionary and prints to the console
```

### Fetch data  from Coptic Orthodox liturgical chant digital collection ---> Processes into a Python dictionary ---> Prints the **"Title"** and **"Description"** of each collection

```python
import requests # Import the requests library for making HTTP requests

response = requests.get("https://www.loc.gov/collections/coptic-orthodox-liturgical-chant/?fo=json") # Send a GET request to the URL with JSON format option
data = response.json()  #Converts the JSON response to a Python dictionary

for collection in data["results"]:
    print(f"Title: {collection['title']}") # Print the title of the collection
    print(f"Description: {collection.get('description', '')}")  # Try to print the description, providing an empty string if not found
    print("-" * 20) # Print a line of hyphens for separation
```

## Checking Status Code

In this code, we first import the `requests` library. This library allows us to send HTTP requests. Then, we send a GET request to the specified URL associated with the 'Coptic Orthodox Liturgical Chant' collection. Finally, we print the status code. The status code is a number that indicates the status of the request. For example, a status code of 200 indicates a successful request.


```python
import requests
response = requests.get("https://www.loc.gov/collections/coptic-orthodox-liturgical-chant/")
print(response.status_code)
```
